# In this section we will be assembling data

```python
import pandas as pd

df1 = pd.read_csv("./data/concat_1.csv")
df2 = pd.read_csv("./data/concat_2.csv")
df3 = pd.read_csv("./data/concat_3.csv")
```

```python
df1
```

```python
df2
```

```python
df3
```

```python


row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
row_concat.loc[0]
```

```python
row_concat.iloc[0]
```

```python
new_row = pd.Series(['n1', 'n2', 'n3', 'n4'])
pd.concat([df1, new_row])
```

```python
# note the double bracket in first argument
new_row_1 = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A', 'B', 'C', 'D'])
new_row_1
pd.concat([df1, new_row_1])
```

```python
# concatenation is similar {only thing that needs to be done is} specify axis=1
cols_concat = pd.concat([df1, df2, df3], axis=1)
cols_concat
```

```python
cols_concat['A']
```

```python
# rename the row index {this is where difference between loc and iloc} will start taking effect
df1.index = [0, 2, 5, 7]
df1
```

```python
# this is useful in timeseries stuff
pd.concat([df1, df2], axis=1)
```
